**OBS: Para abrir os arquivos, o encoding necessário é "cp1250"**

```python
import pandas as pd
import os

year = "2024"
month = "04"
licitacoes_csv_file_path = os.getcwd() + f"\\data\\{year}{month}_Licitacoes\\{year}{month}_Licitação.csv"
# Reading with proper encoding ans separator
df = pd.read_csv(licitacoes_csv_file_path, encoding="cp1250", sep=';')
print(df.shape)
df.iloc[:3]
```
